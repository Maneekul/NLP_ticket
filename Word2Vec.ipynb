```python
import pandas as pd
from nltk.corpus import stopwords as en_stopwords
import nltk
from pythainlp.corpus import thai_stopwords

nltk.download('stopwords')

stop_words_th = set(thai_stopwords())
stop_words_en = set(en_stopwords.words('english'))

# Add Thai characters 'ก' to 'ฮ' to the Thai stop words set
thai_alphabet = [chr(i) for i in range(ord('ก'), ord('ฮ') + 1)]
# List of Thai vowels and tone marks
thai_vowels_and_tone_marks = ['ะ', 'ั', 'า', 'ำ', 'ิ', 'ี', 'ึ', 'ื', 'ุ', 'ู', 'เ', 'แ', 'โ', 'ใ', 'ไ', 'ๅ', '็', '่', '้', '๊', '๋', '์', 'ํ']

stop_words_th.update(thai_alphabet)
stop_words_th.update(thai_vowels_and_tone_marks)

combined_stop_words = stop_words_th.union(stop_words_en)

# Function to remove stop words
def remove_stop_words(tokens, stop_words):
    return [token for token in tokens if token.lower() not in stop_words]
```

```python
my_df['words_list'] = my_df['words_list'].apply(lambda tokens: remove_stop_words(tokens, combined_stop_words))
```

```python
import re
import numpy as np
import pandas as pd
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Define the text cleaning function
def cleanText(text):
    text = str(text)
    text = re.sub(r'([^ก-๙a-zA-Z\s])', '', text)
    stop_word = list(thai_stopwords())
    sentence = word_tokenize(text, engine="longest")
    result = [word for word in sentence if word not in stop_word and " " not in word]
    return ",".join(result)

def tokenize(d):
    result = d.split(",")
    result = list(filter(None, result))
    return result

new_text = [cleanText(txt) for txt in my_df['words_list']]

vectorizer = CountVectorizer(tokenizer=tokenize, token_pattern=None)
transformed_data = vectorizer.fit_transform(new_text)
feature_names = vectorizer.get_feature_names_out()
word_counts = np.ravel(transformed_data.sum(axis=0))
keyword_df = pd.DataFrame(list(zip(feature_names, word_counts)), columns=['word', 'count'])
top_keywords = keyword_df.sort_values(by=['count'], ascending=False).head(10)
print(top_keywords)
```

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Combine training data (using only 'text' column from train_corpus)
combined_training = train_corpus['text'].tolist()

# Train Word2Vec model
train_data = [text.split() for text in combined_training]
w2v_model = Word2Vec(train_data, vector_size=300, min_count=2, window=5, sg=1, workers=4)

# Function to get Word2Vec embeddings
def get_embedding_w2v(doc_tokens):
    embeddings = []
    if len(doc_tokens) < 1:
        return np.zeros(300)
    else:
        for token in doc_tokens:
            if token in w2v_model.wv.key_to_index:  # Check if token is in vocabulary
                embeddings.append(w2v_model.wv.get_vector(token))
            else:
                embeddings.append(np.random.rand(300))
        return np.mean(embeddings, axis=0)

# Apply embeddings to test corpus
test_corpus['vector'] = test_corpus['text'].apply(lambda x: get_embedding_w2v(x.split()))

# Function for information retrieval ranking
def ranking_ir(query):
    vector = get_embedding_w2v(query.split())

    documents = test_corpus[['labels', 'text']].copy()
    documents['similarity'] = test_corpus['vector'].apply(lambda x: cosine_similarity(np.array(vector).reshape(1, -1), np.array(x).reshape(1, -1)).item())
    documents.sort_values(by='similarity', ascending=False, inplace=True)
    return documents.head(10).reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Load train and test data
train_corpus_token = pd.read_csv('/content/drive/MyDrive/Data_wok/train_df.csv')
test_corpus_token = pd.read_csv('/content/drive/MyDrive/Data_wok/test_df.csv')

# Combine training data (using only 'text' column from train_corpus)
combined_training_token = train_corpus_token['tokenized'].tolist()

# Train Word2Vec model
train_data = [text.split() for text in combined_training_token]
w2v_model = Word2Vec(train_data, vector_size=300, min_count=2, window=5, sg=1, workers=4)

# Function to get Word2Vec embeddings
def get_embedding_w2v_token(doc_tokens):
    embeddings = []
    if len(doc_tokens) < 1:
        return np.zeros(300)
    else:
        for token in doc_tokens:
            if token in w2v_model.wv.key_to_index:  # Check if token is in vocabulary
                embeddings.append(w2v_model.wv.get_vector(token))
            else:
                embeddings.append(np.random.rand(300))
        return np.mean(embeddings, axis=0)

# Apply embeddings to test corpus
test_corpus_token['vector'] = test_corpus_token['tokenized'].apply(lambda x: get_embedding_w2v_token(x.split()))

# Function for information retrieval ranking
def ranking_ir_token(query):
    vector = get_embedding_w2v_token(query.split())

    documents_token = test_corpus_token[['labels', 'tokenized']].copy()
    documents_token['similarity'] = test_corpus_token['vector'].apply(lambda x: cosine_similarity(np.array(vector).reshape(1, -1), np.array(x).reshape(1, -1)).item())
    documents_token.sort_values(by='similarity', ascending=False, inplace=True)
    return documents_token.head(10).reset_index(drop=True)
```
